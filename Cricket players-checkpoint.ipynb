```python
#Filter dataset for india and australia
df1_filtered = df1[
    (df1['Team 1'].isin(['Australia', 'India'])) &
    (df1['Team 2'].isin(['Australia', 'India']))
]

# Converting categorical values to encoded values
x = df1_filtered.drop(['Winner','Unnamed: 0','Scorecard','Match Date','Host_Country'],axis =1)
y = df1_filtered['Winner']
x_filtered = pd.get_dummies(x,drop_first = True)
y_filtered = pd.get_dummies(y,drop_first = True)
```

```python
# Splitting data into training and test set
#import relevant libraries
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_filtered,y_filtered,test_size = 0.3,random_state = 0)
```

```python
from xgboost import XGBClassifier

classifier = XGBClassifier(learning_rate=0.25,n_estimators=500,objective='binary:logistic')

classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import roc_auc_score

display(roc_auc_score(y_test,y_pred))

display(x_filtered.head())
```

```python
def display_prediction(x_filtered, venue_name, venue_condition):
    x_current = x_filtered[(x_filtered['Team 1_India'] == 1) & (x_filtered['Venue_Team1_Home'] == 1) & venue_condition]
    result = classifier.predict(x_current).mean()
    display(f"{venue_name}: {result}")

#venue conditions
venue_conditions = {
    'Visag': (x_filtered['Ground_Visakhapatnam'] == 1),
    'Bangalore': (x_filtered['Ground_Bengaluru'] == 1),
    'Hyderabad': (x_filtered['Ground_Hyderabad'] == 1),
    'Nagpur': (x_filtered['Ground_Nagpur'] == 1),
    'New Delhi': (x_filtered['Ground_New Delhi'] == 1) | (x_filtered['Ground_Delhi'] == 1)
}

# Display predictions for each venue
for venue_name, condition in venue_conditions.items():
    display_prediction(x_filtered, venue_name, condition)
```
